```r
# Question 1 & 2
# My guess is that we are forgeting in most regressions there is an constant, i.e. y = alpha + betax
# we should therefore add the alpha to each element of vector X, what will make beta a vector

require(MASS)

beta = 0.5
SIGMA = matrix(c(1,beta,beta,1), ncol=2)
MU = c(2.0, 1.0)

N = 50

out <- mvrnorm(N, mu = MU, Sigma = SIGMA)

y <- out[, 2]
X <- cbind(rep(1,N),out[, 1]) #here is the change - question: it didn't work with matrix((1:N), out[,1]) - why? 

XT = t(X)
XTX = XT%*%X
invXTX = solve(XTX)
XTy = XT%*%y
beta = invXTX %*% XTy
beta [1] #here is alpha
beta [2] #here is beta

plot(out)
abline(lm(out[,2]~out[,1]), col="red")
abline(a=beta[1], b=beta[2], col="blue")
```

```r
# Question 3
# Now there are 3 variables so we need to define another mu

MU = c(2.0, 1.0, 3.0)

#and covariances!

beta1 = 0.2 #=cov12=cov21
beta2 = 0.3 #=cov13=cov31
beta3 = 0.4 #=cov23=cov32

SIGMA = matrix(c(1, beta1, beta2, beta1, 1, beta3, beta2, beta3, 1), ncol=3)

N = 50

out2 <- mvrnorm(N, mu = MU, Sigma = SIGMA)

y <- out2[, 3]
X <- cbind(rep(1,N),out2[, c(1, 2)])  # need to add a column to X obviously

XT = t(X)
XTX = XT%*%X
invXTX = solve(XTX)
XTy = XT%*%y
beta = invXTX %*% XTy
beta [1] 
beta [2]
beta [3]

```
