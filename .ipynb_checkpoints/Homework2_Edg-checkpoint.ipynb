```r
# Question 2 & 3 : you make all loops depend on the same random event (runif) : 
# we see on the plots the similarity when taking very close values of p 

w0 = 100
p = c(0.51,0.49)
I = 2
T = 1000 

w1 = rep(0, T+1)
w2 = rep(0,T+1)

w1[1] = w0
w2[1] = w0

##### Comment
# Nice, but hopefully you can see now how to add the loop here
# after looking at the correction. Whenever you see code
# repeated, you can usually use a loop, but you need a matrix
# to store your data
for(t in 1:T){
    u = runif(1)
    if(u <= 1 - p[1]){        
        w1[t+1] = w1[t] - 1
    } else {
        w1[t+1] = w1[t] + 1
    }
    if(u <= 1 - p[2]){
        w2[t+1] = w2[t] - 1
    } else {
        w2[t+1] = w2[t] + 1
    }
    
}
time = (0:T+1)
plot(time,w1,type="l",col="red")
lines(time,w2,col="green")
```

```r
# Question 4 : now I have to loop different w's, so I'll create a vector where I will store these w's and then take its mean 


w0 = 100
p = 0.4
T = 1000
A = 10

W = c(0,A)

for (i in 1:A){
    w1 = c(0,T+1)
    w1[1] = w0
    for(t in 1:T){
        u = runif(1)
        if(u <= 1 - p){        
            w1[t+1] = w1[t] - 1
        } else {
            w1[t+1] = w1[t] + 1
        }
    W[i] = w1[T+1]
    }
}

Average = mean(W)
Average
```
